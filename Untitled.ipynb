```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import time
```

```python
def data_extractor(driver):
#     time.sleep(2)
    name=age=bio=edu=job=gender=lives=distance=""
    soup=BeautifulSoup(driver.page_source,'lxml')
    try:
        name = soup.find(class_='Fz($xl) Fw($bold) Flxs($flx1) Flw(w) Pend(8px)').getText()
    except:
        name=""
    # print('Name : {}'.format(name))
    try:
        age = soup.find(class_='Whs(nw) Fz($l)').getText()
    except:
        age =""

    try:
        bio=soup.find(class_='P(16px) profileCard__bio Ta(start) Us(t) C($c-secondary) BreakWord Whs(pl) Fz($ms)').getText()

    except:
        bio=""

#     try:
    
#     edu=job=gender=lives=distance=""
    rows = soup.find_all(class_='Row')
    print(len(rows))
    row_len = len(rows)
    print('---')
    if row_len!=0:
        for r in rows:
            if 'M11.87 5.026L2.186 9' in r.find('path')['d']:
                # print('edu')
                edu = r.getText()
#                 print(r.getText())
            elif 'M7.15 3.434h5.7V1.45' in r.find('path')['d']:
                # print('job')
                job = r.getText()
#                 print(r.getText())      
            elif 'M17.507 13.032c1' in r.find('path')['d']:
                # print('gender')
                gender = r.getText()
                # print(r.getText())     
            elif r.find(class_='Us(t) Va(m) D(ib) My(2px) NetWidth(100%,20px) C($c-secondary) Ell'):
                # print('lives')
                lives = r.getText()
                # print(r.getText())
            else:
                # print('distance')
                distance = r.getText()
            # print(r.getText())
#     except Exception as e:
#         edu=job=gender=lives=distance=""
#         print(e)
#     print(name,age,bio,edu,job,gender,lives,distance)
    print("name : {}\n,age : {}\n,bio : {}\nedu:{}\n,job : {}\n,gender : {}\n,lives : {}\n,distance : {}\n".format(name,age,bio,edu,job,gender,lives,distance))
    name=age=bio=edu=job=gender=lives=distance=""


    print('\n\n')
```
